```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1001)

# Logic check: Your legend mentioned +2, so I added it to y2 to match!
y1 = 3**(x-1)
y2 = 2**(-x-2)

# We use the comma after gp1/gp2 because plt.plot returns a list 
# This "unpacks" the list so we can use the line objects in the legend
gp1, = plt.plot(x, y1)
gp2, = plt.plot(x, y2)

# Fix: 'labels' was misspelled as 'lables'
# Fix: 'handles' should typically be in a list [] or tuple ()
plt.legend(handles=[gp1, gp2], labels=(r'$y=3^x$', r'$y=2^{-x-2}$'))

plt.show()
```

```python
x1 = np.linspace(0.1,4,401)
y1 = np.log2(x1)
y2 = np.log2(x1 + 1) - 1
gp1, = plt.plot(x1,y1)
gp2, = plt.plot(x1,y2)
plt.legend(handles =[gp1,gp2], labels = (r'$y=log_2(x)$',r'$y=log_2(x + 1) - 1$'))
plt.show()
```

```python
A = np.array([[1,-4],[-4,1]])
B = np.array([[2,-8],[-8,2]])
print("A + B = ", A + B)
```

```python
print("A - (1/2)B = ", A - (1/2)*B)
```

```python
print("A * B = ", A * B)
```

```python
detA = np.linalg.det(A)
Ainv = np.linalg.inv(A)
detB = np.linalg.det(B)
```

```python
print("DET A = ", detA)
```

```python
print("INV A = ", Ainv)
```

```python
print("DET B = ", detB)
```

```python
import sympy as sp

x = sp.Symbol('x')
fx = x**2020 + 3 * x * x * x - 1
```

```python
sp.diff(fx,x)
```

```python
sp.diff(sp.diff(fx,x),x)
```

```python
# Sigmoid fun

m = sp.Symbol('m')
sig = 1/(1 + sp.exp(-m))
sp.diff(sig,m)
```
